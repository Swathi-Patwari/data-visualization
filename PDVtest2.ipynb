```python
#5. line graph of milk production in 5 years
import matplotlib.pyplot as plt
from pandas import DataFrame
import pandas as pd

data = pd.read_csv('milk_production1.csv', index_col = False)
data["y2011"] =(data["Cow Milk-2010-11"] + data["Boffalo Milk-2010-11"] + data["Goat Milk-2010-11"])
data["y2012"] =(data["Cow Milk-2011-12"] + data["Boffalo Milk-2011-12"] + data["Goat Milk-2011-12"])
data["y2013"] =(data["Cow Milk-2013-14"] + data["Boffalo Milk-2013-14"] + data["Goat Milk-2013-14"])
data["y2014"] =(data["Cow Milk-2014-15"] + data["Boffalo Milk-2014-15"] + data["Goat Milk-2014-15"])
data["y2015"] =(data["Cow Milk-2015-16"] + data["Boffalo Milk-2015-16"] + data["Goat Milk-2015-16"])

#plot data
plt.plot(data['State/ UT Name'], data['y2011'], linestyle="dotted", marker='^', color="red", label = 'year2011')

plt.plot(data['State/ UT Name'], data['y2012'], linestyle="solid", marker='o', color="green", label = 'year2012')

plt.plot(data['State/ UT Name'], data['y2013'], linestyle="dashed", marker='<', color="blue", label = 'year2013')

plt.plot(data['State/ UT Name'], data['y2014'], linestyle="dashed", marker='<', color="blue", label = 'year2014')

plt.plot(data['State/ UT Name'], data['y2015'], linestyle="dashed", marker='<', color="blue", label = 'year2015')

#labels
plt.xlabel("state/ UT")
plt.ylabel("for particular year milk production in Litres")

# add legend
plt.legend()

#title
plt.title("Milk Poduction Analysis")

#show plot
plt.show()
```
